```python
class FullTimeEmployee(Employee):
    def __init__(self, e_name, emp_addr, veh, salary):
        
        self.__salary  = salary
        self.__vehicle = veh
       
        self.__emp_det = Employee(e_name, emp_addr, veh) 
        
    
   # accessor functions/methods 
    def get_salary(self):     
        return self.__salary
      
  
    def get_emp_det(self):
        return self.__emp_det
    
    def get_vehicle(self):
        return self.__vehicle
    
    def get_emp_name(self):
        return self.__emp_det.get_emp_name()
    
    def get_vehicle_year(self):
        return self.__emp_det.get_vehicle_year()
    
    def get_vehicle_mileage(self):
        return self.__emp_det.get_vehicle_mileage()
    
   
    #mutator functions/methods
    
    def set_salary(self,salary):     
        self.__salary = salary    
        

   
    
    def compute_compensation(self):
        if self.__salary <= 45000:
            tax_rate = 0.18
            tax = 0.18*self.__salary
            return ((self.__salary-tax)/52)
        
        elif self.__salary > 45000 and self.__salary <= 82000:
            tax_rate1 = 0.18
            tax_rate2 = 0.28
            tax = (0.18*45000) + (0.28*(self.__salary-45000))
            return (self.__salary-tax)/52
        
        elif self.__salary > 82000:
            tax_rate1 = 0.18
            tax_rate2 = 0.28
            tax_rate3 = 0.33
            tax = (0.18*45000) + (0.28*(82000-45000)) + (0.33*(self.__salary-82000))
            return ((self.__salary-tax)/52)
    
    
    
    def compute_reimbursement(self, annual_exp):  
        if annual_exp <= 10000:
            reimburse = annual_exp / 52
            return reimburse
        
        elif annual_exp > 10000:
            reimburse = (10000 + (0.5 * (annual_exp - 10000)))/52   
            return reimburse
        
        
        
        
    def __str__(self):
        emp_det = '{}'.format(self.__emp_det.__str__())
        emp_det += '\nSalary: {:0.2f}'.format(self.__salary)
        return "\nDetails of this Full Time Employee are:" + emp_det
    
    
```

```python
class HourlyEmployee(Employee):
    def __init__(self, e_name, emp_addr, vehicle, num_hrs, rate):
        
        self.__num_hrs_wrkd  = num_hrs
        self.__hourly_rate = rate
        self.__emp_salary = self.__hourly_rate*self.__num_hrs_wrkd 
        
        self.__emp_det = Employee(e_name, emp_addr, vehicle) 
    
    
    
    # accessor functions/methods
    
    def get_hours_worked(self):     
        return self.__num_hrs_wrkd
        
    def get_hourly_rate(self):     
        return self.__hourly_rate
    
    def get_salary(self):
        return self.__emp_salary 

    def get_emp_det(self):
        return self.__emp_det
    
    def get_emp_name(self):
        return self.__emp_det.get_emp_name()
    
    def get_vehicle_year(self):
        return self.__emp_det.get_vehicle_year()
    
    def get_vehicle_mileage(self):
        return self.__emp_det.get_vehicle_mileage()
    
   
    #mutator functions/methods
    
    def set_hours_worked(self, hours_worked):     
        self.__num_hrs_wrkd = hours_worked
        
    def set_hourly_rate(self, hourly_rate):     
        self.__hourly_rate = hourly_rate
    
  
        
        
    
    def compute_compensation(self):
        if self.__num_hrs_wrkd <= 40:
            compensation = self.__num_hrs_wrkd*self.__hourly_rate
            return compensation
        
        elif self.__num_hrs_wrkd > 40:
            compensation1 = (40 * self.__hourly_rate)
            compensation2 = (self.__num_hrs_wrkd - 40)*self.__hourly_rate*1.8
            compensation = compensation1 + compensation2   
            return compensation
        
        
        
    def compute_reimbursement(self, weekly_exp):  
        if weekly_exp <= 100:
            reimburse = weekly_exp
            return reimburse
        
        elif weekly_exp > 100:
            reimburse = 100  
            return reimburse
    

    def __str__(self):
        emp_det = '{}'.format(self.__emp_det.__str__())
        emp_det += '\nHours Worked: {}; Hourly Rate: {}'.format(str(self.__num_hrs_wrkd), str(self.__hourly_rate))
        return "\nDetails of this Hourly Employee are:" + emp_det
    
 
```

```python
class Consultant(Employee):
    def __init__(self, e_name, emp_add, veh, num_hrs, proj):
        
        if proj<=0 or proj>3:
            print('Invalid value for Project Type')
            sys.exit(0)
            
        self.__num_hrs_wrkd  = num_hrs
        self.__proj_type = proj
        
        self.__emp_det = Employee(e_name, emp_add, veh) 
       
    
    # accessor functions/methods
    
    def get_hours_worked(self):     
        return self.__num_hrs_wrkd
        
    def get_project_type(self):     
        return self.__proj_type
    
       
    def get_emp_det(self):
        return self.__emp_det
    
    def get_emp_name(self):
        return self.__emp_det.get_emp_name()
    
    def get_vehicle_year(self):
        return self.__emp_det.get_vehicle_year()
    
    def get_vehicle_mileage(self):
        return self.__emp_det.get_vehicle_mileage()
    
    
    #mutator functions/methods
    
    def set_hours_worked(self, num_hrs_wrkd):     
        self.__num_hrs_wrkd = num_hrs_wrkd
        
    def set_project_type(self, proj_type):     
        self.__proj_type = proj_type
    
    def set_emp_det(self, emp_det):
        self.__emp_det = emp_det
    
  
    
    def compute_compensation(self):
        if self.__proj_type == 1:
            return 55.00*self.__num_hrs_wrkd
        
        elif self.__proj_type == 2:
            return 70.00*self.__num_hrs_wrkd
        
        elif self.__proj_type == 3:
            return 85.00*self.__num_hrs_wrkd
        
    
    
    def compute_reimbursement(self, weekly_exp):  
        if self.__proj_type == 1:
            return 1.00*weekly_exp
        
        elif self.__proj_type == 2:
            return 0.90*weekly_exp
        
        elif self.__proj_type == 3:
            return 0.80*weekly_exp
        
    

    def __str__(self):
        emp_det = '{}'.format(self.__emp_det.__str__())
        emp_det += '\nHours Worked: {}; Project Type: {}'.format(str(self.__num_hrs_wrkd), self.__proj_type)
        return "\nDetails of this Consultant are:" + emp_det
    
```

```python
#complete this get_vehicle_input() function to prepare for the main menu application
#this function is used to ask for vehicle information and then return them: make,model,year,mileage
#this function doesn't take arguement(s)
#no need to handle exceptions of make input and model input
#handle exceptions to make sure year is a 4-digit positive number between 1900 and 2020
#handle exceptions to make sure mileage is a positive number

def get_vehicle_input():
    vehiclemake = input('Enter the vehicle make: ')
    vehiclemodel = input('Enter the vehicle model: ')

    
    while True:
        try:
            vehicleyear = int(input('Enter the year of manufacture (yyyy): '))
            if vehicleyear < 1900 or vehicleyear > 2020:
                    raise Exception    # We raise our own error

        except Exception:
            print('\nPlease enter an integer value for year in the format of yyyy between 1900 and 2020.')
            pass
            

        else:
            while True:
                try:
                    vehiclemileage = int(input('Enter the mileage: '))
                    if vehiclemileage <= 0:
                        raise ValueError
                except ValueError:
                    print('\nPlease enter a valid input!')
                    pass

                else:
                    return (vehiclemake, vehiclemodel, vehicleyear, vehiclemileage)  



    
    
```

```python
#complete this read_file_data() function to prepare for the main menu application
#this function is used to read information from an existing databse/binary file, which is empdata.dat in the final exam
#this function doesn't take argument(s)
#this function returns a list containing all employee objects stored in the database; each employ object is an element in the list; the list will also be used and updated later
#need to handle exceptions if file doesn't exist

import pickle
def read_file_data():
    
    emplist=[]
    
    try:

        fb = open('empdata.dat', 'rb') 
       
        while(True):          # this while loop will be always entered since the conditional expression will always be True
                emplist.append(pickle.load(fb))
                
                
    except FileNotFoundError:         #code execution will transfer here if "File not found"
            print("Wrong file or file path")
    
    except EOFError: #code execution will transfer here when the EOF condition is encountered
        pass
  
    fb.close()
    return emplist
```

```python
#complete option 4 related code here
#the function is used to search for an employee by name and the searching is NOT case sensitive
#if there are some matching cases, print the information of all matched employees
#if no matchng cases, print "There is no employee matching the name you entered."
#this function takes one argument, which is the emp_lst list (i.e. the list created previously which is used to store employee objects)
#this function also returns a list of objects of the matched employees
#no need to handle exceptions

def run_option4(my_lst):
   
    emp_name = input("\nPlease enter the name of the employee you want to search:")
    
    
    matched_emp_list = []
    count=0
    
    
    for i in range(0,len(my_lst)):  
        if emp_name.lower() == my_lst[i].get_emp_name().lower():
            count+=1
            matched_emp_list.append(my_lst[i])
         
    if count>0:
        print("==========================================================================")        
        print("Below is the information of all employees that match the name you entered.")
        print("==========================================================================")  
        for i in matched_emp_list:
            print(i)
           
        
    elif count==0:
            print('There is no employee matching the name you entered.')
        
        
   
    
    return matched_emp_list
```

```python
#complete option 5 related code here
#the function is used to show some basic statistics. They are: 
#(1) the number of employees stored in the database
#(2) the highest weekly compensation
#(3) the mean weekly compensation
#(4) the number of employees who have a vehicle with over 100,000 mileage
#this function takes one argument, which is the emp_lst list (i.e. the list created previously which is used to store employee objects)
#this function doesn't return anything
#no need to handle exceptions

def run_option5(my_lst):
    
    print("\n=========================================================")
    print("Below is the statistics of all employees in this database.")
    print("==========================================================")
    
    
    compensa = 0
    compensation_amount=0
    veh_mileage = 100000
    tempname1 = []
    
  
    
    for i in range(0, len(my_lst)):
        
        #Highest weekly compensation
        if compensa < my_lst[i].compute_compensation():
            compensa = my_lst[i].compute_compensation()
        
        #Vehicle Mileage
        if veh_mileage < my_lst[i].get_vehicle_mileage():
            veh_mileage = my_lst[i].get_vehicle_mileage()
            name = my_lst[i].get_emp_name()
            
            if veh_mileage == my_lst[i].get_vehicle_mileage() and name != my_lst[i].get_vehicle_mileage():
                tempname1.append(my_lst[i].get_emp_name())  
        
        #Mean Weekly Compensation
        compensation_amount += my_lst[i].compute_compensation()
    mean_compensa = compensation_amount/len(my_lst)
            

    print("There are {} employees stored in this employee database.".format(len(my_lst)))
    print("The highest weekly compensation is ${:0.2f}.".format(compensa))
    print("The mean weekly compensation is: ${:0.2f}.".format(mean_compensa))
    print("The number of employees who have a vehicle with over 100,000 mileage is {}.".format(len(tempname1)))
    
    
    
    
```
